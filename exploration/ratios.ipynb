```python
import pandas as pd

# Ratios for a lot of different dataframes

df = pd.read_csv('../data/metadata/fixed_metadata.csv')

df = df.sort_values(["lesion_id"]) 
df = df[~df["gender"].isna()] # Removing data without gender entry
df = df.drop_duplicates("lesion_id", keep = "first") # Dropping duplicates, but keeping the first occurence of the lesion_id
df['is_cancerous'] = df['diagnostic'].apply(lambda x: any(cancer in x for cancer in ['SCC', 'BCC', 'MEL'])).astype(int)
df_once_augmented = pd.read_csv('../data/metadata/once_augmented.csv')

print(df.shape)

print("Ratio of cancer/non-cancerous before augment for full df: ", df[df['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , df[df['is_cancerous'] == 0]['is_cancerous'].value_counts())

print("Ratio of cancer/non-cancerous after augment for full df: ", df_once_augmented[df_once_augmented['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , df_once_augmented[df_once_augmented['is_cancerous'] == 0]['is_cancerous'].value_counts())


testdf = pd.read_csv('../data/cnn/cnn_splitted_data_once_augmented/m_f_ca_nc_test_0.csv')
testlist = testdf['img_id'].to_list()

df = df[~df['img_id'].isin(testlist)]
print(df.shape)
old_train_df = pd.read_csv('../old/data_split/m_f_ca_nc_train_0_0.00_0.csv')
old_val_df = pd.read_csv('../old/data_split/m_f_ca_nc_val_0_0.00_0.csv')
old_train_val_df = pd.concat([old_train_df, old_val_df])


df_once_augmented = df_once_augmented[~df_once_augmented['img_id'].isin(testlist)]

new_train_df = pd.read_csv('../data/cnn/cnn_splitted_data_once_augmented/m_f_ca_nc_train_1_0.00_1.csv')
new_val_df = pd.read_csv('../data/cnn/cnn_splitted_data_once_augmented/m_f_ca_nc_val_1_0.00_1.csv')
new_train_val_df = pd.concat([new_train_df, new_val_df])


print("Ratio of cancer/non-cancerous before augment for full df without one test: ", df[df['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , df[df['is_cancerous'] == 0]['is_cancerous'].value_counts())
# Ratio is 781:294
print("Ratio of cancer/non-cancerous before augment for random train: ", old_train_df[old_train_df['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , old_train_df[old_train_df['is_cancerous'] == 0]['is_cancerous'].value_counts())
# Ratio is 247:85

print("Ratio of cancer/non-cancerous after augment for full df without one test: ", df_once_augmented[df_once_augmented['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , df_once_augmented[df_once_augmented['is_cancerous'] == 0]['is_cancerous'].value_counts())
# Ratio is 781:640
print("Ratio of cancer/non-cancerous after augment for random train: ", new_train_df[new_train_df['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , new_train_df[new_train_df['is_cancerous'] == 0]['is_cancerous'].value_counts())
# Ratio is 236:192

print("Ratio of cancer/non-cancerous after augment for random train/val: ", new_train_val_df[new_train_val_df['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , new_train_val_df[new_train_val_df['is_cancerous'] == 0]['is_cancerous'].value_counts())

print("Ratio of cancer/non-cancerous before augment for random train/val: ", old_train_val_df[old_train_val_df['is_cancerous'] == 1]['is_cancerous'].value_counts(), ":" , old_train_val_df[old_train_val_df['is_cancerous'] == 0]['is_cancerous'].value_counts())


```
