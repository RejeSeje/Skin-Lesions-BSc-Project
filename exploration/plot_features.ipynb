```python
import pandas as pd
```

```python
data = pd.read_csv("../../BSc-Project/data/metadata/augmented_master_data.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'merged_df' is your DataFrame
sns.set(style="white")  # Set the style of the plot to be white

palette = {
    'BCC': 'pink',    # Example color for BCC
    'SCC': 'orchid',   # Example color for SCC
    'MEL': 'magenta',  # Example color for MEL
    'ACK': 'darkgreen', # Example for ACK
    'SEK': 'lightgreen', # Example for SEK
    'NEV': 'green'   # Example for NEV
    # Add other diagnostics and their colors as needed
}
#F2, F11, sat_var, blue_veil_pixels, avg_green_channel, mean_asymmetry, F1, average_hue, compactness_x and dom_hue

# Create a scatter plot with marginals. Make sure to pass the 'palette' parameter here
g = sns.jointplot(data=data, x="F2", y="F11", hue="diagnostic", palette=palette, kind="scatter", height=10, ratio=4, marginal_ticks=True)

# Add legends and improve layout
plt.subplots_adjust(top=0.95)  # Adjust the top to make space for the title if necessary
g.fig.suptitle('Scatter Plot of F2 vs. F11 with Diagnostics Color Coding')  # Add a title to the figure

# Show plot
plt.show()
```
